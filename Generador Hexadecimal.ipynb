```python
import tkinter as tk
import random
import re

def generar_contrasena():
    try:
        longitud = int(input_longitud.get())
        caracteres = "123456789ABCDEF"
        contrasena = ''.join(random.choice(caracteres) for _ in range(longitud))
        lbl_contrasena.config(text=contrasena)

        if es_contrasena_segura(contrasena):
            lbl_seguridad.config(text="Segura")
        else:
            lbl_seguridad.config(text="No segura")

    except ValueError:
        lbl_contrasena.config(text="Error: Longitud inválida")


def es_contrasena_segura(contrasena):
    tiene_mayuscula = re.search(r'[A-Z]', contrasena)
    tiene_minuscula = re.search(r'[a-z]', contrasena)
    tiene_digito = re.search(r'\d', contrasena)
    tiene_caracter_especial = re.search(r'[!@#$%^&*]', contrasena)

    return (
        tiene_mayuscula and
        tiene_minuscula and
        tiene_digito and
        tiene_caracter_especial
    )


# Colores
color_primario = "#3498DB"    # Azul claro
color_secundario = "#E67E22"  # Naranja
color_fondo = "#ECF0F1"       # Gris claro
color_texto = "#2C3E50"       # Azul oscuro

# Fuentes
fuente_titulo = ("Arial", 24, "bold")
fuente_etiqueta = ("Arial", 14)
fuente_boton = ("Arial", 14, "bold")
fuente_contrasena = ("Arial", 16, "bold")

# Ventana
ventana = tk.Tk()
ventana.title("Generador de Contraseñas")
ventana.configure(bg=color_fondo)
ventana.geometry("500x400")
ventana.resizable(False, False)

# Título
lbl_titulo = tk.Label(ventana, text="Generador de Contraseñas", bg=color_fondo, fg=color_primario, font=fuente_titulo)
lbl_titulo.pack(pady=(30, 20))

# Marco para la longitud
frame_longitud = tk.Frame(ventana, bg=color_fondo)
frame_longitud.pack()
lbl_longitud = tk.Label(frame_longitud, text="Longitud:", bg=color_fondo, fg=color_secundario, font=fuente_etiqueta)
lbl_longitud.pack(side=tk.LEFT, padx=10)
input_longitud = tk.Entry(frame_longitud, font=fuente_etiqueta, width=5)
input_longitud.pack(side=tk.LEFT)

# Botón para generar
btn_generar = tk.Button(ventana, text="Generar Contraseña", command=generar_contrasena, bg=color_primario, fg=color_fondo, font=fuente_boton, relief="flat", width=20, height=2)
btn_generar.pack(pady=10)

# Contraseña generada
lbl_contrasena = tk.Label(ventana, text="", bg=color_fondo, fg=color_texto, font=fuente_contrasena)
lbl_contrasena.pack()

# Seguridad de la contraseña
lbl_seguridad = tk.Label(ventana, text="", bg=color_fondo, fg=color_texto, font=fuente_etiqueta)
```
